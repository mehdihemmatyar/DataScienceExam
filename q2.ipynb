```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sn#for visuals
sn.set(style="white", color_codes=True)#customizes the graphs
import matplotlib.pyplot as mp #for visuals
%matplotlib inline
import warnings #suppress certain warnings from libraries
warnings.filterwarnings("ignore")

movie_data = pd.read_csv("imdb.csv")#load the dataset
movie_data
```

```python
movie_data.fillna(value=0,axis=1,inplace=True) #delete isnull data
```

```python
movie_data.describe()
```

```python
genres = movie_data.genres
```

```python
movie_data.isnull().sum() # dataset is clear
```

```python
len(genres)
```

```python
import ast
list_genres = [0]*len(genres)
for index , value in enumerate(genres):
    try:
        list_genres[index] = ast.literal_eval(value)
    except:
        continue
```

```python
list_film = list()
for i in list_genres:
    try:
        for j in i:
            list_film.append(j['name'])
    except:
        continue
        
        
    
```

```python
list_film 
```

```python
from collections import Counter
Counter(list_film).most_common() # Answer the first part of the question (1)
```

```python
companies = movie_data.production_companies
```

```python
list_companies = [0]*len(companies)
for index , value in enumerate(companies):
    try:
        list_companies[index] = ast.literal_eval(value)
    except:
        continue
```

```python
list_company = list()
for i in list_companies:
    try:
        for j in i:
            list_company.append(j['name'])
    except:
        continue
        
```

```python
Counter(list_company).most_common() #Answer the third part of the question(3)
```

```python
movie_data['revenue/budget'] = movie_data['revenue'] / movie_data['budget']
```

```python
for i in range(3000): #Some movies have a zero budget, which can lead to split errors
    if (movie_data['revenue/budget'][i] == float('inf')):
        movie_data['revenue/budget'][i] = 0
```

```python
movie_data['revenue/budget'] #celar ratio of revenue / budget
```

```python
genres_income = dict.fromkeys(list(set(list_film)), 0)
for index , i in enumerate(list_genres):
    try:
        for j in i:
            genres_income[j['name']] += movie_data['revenue/budget'][index]
    except:
        continue
```

```python
Counter(genres_income).most_common() #Answer the second  part of the question(2)
```
