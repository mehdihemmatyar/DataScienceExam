```
import math
def cost(p1):
  c = 0
  x_bank = [2000, 4200, 3500, 1000, 700]
  y_bank = [2500, 4000, 500, 4200, 500]
  min_dist = 10000
  for p2 in zip(x_bank , y_bank):
    distance = math.sqrt( ((p1[0]-p2[0])**2)+((p1[1]-p2[1])**2))
    if distance <= min_dist:
      min_dist = distance
  if min_dist < 500:
    c = 1
  elif min_dist< 1000:
    c = 3
  else:
    c = 5
  return c
```
